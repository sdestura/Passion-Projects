```python
import string
import nltk
from nltk.corpus import stopwords
all_stopwords = stopwords.words('english')
#if you want to view all stopwords from the NLTK library,
#uncomment the print statement
#print(all_stopwords)

#you can append more stopwords as you see fit
#keep in mind this may affect your results
all_stopwords.append('like')
all_stopwords.append('um')
all_stopwords.append('okay')
all_stopwords.append('im')
all_stopwords.append('yeah')
all_stopwords.append('think')
all_stopwords.append('go')
all_stopwords.append('would')
all_stopwords.append('know')
all_stopwords.append('thats')
all_stopwords.append('dont')
all_stopwords.append('right')
all_stopwords.append('uh')
all_stopwords.append('mm')
all_stopwords.append('hm')
all_stopwords.append('hmm')
all_stopwords.append('oh')
all_stopwords.append('thing')
all_stopwords.append('things')
all_stopwords.append('kind')
all_stopwords.append('well')
all_stopwords.append('see')
```

<h2> Building our Class & Methods üõ† </h2>

```python
class analysedText(object):
    
    def __init__ (self, text):
        #removing timestamps
        #check each line in the list, and filter for those that start with an alpha character. 
        #every line that we match is used to build the string using formattedText += line 
        formattedText = ''
        for line in text:
            if line[0].isalpha():
                formattedText += line
        #print('Without timestamps:', formattedText)
                
        #formatting text by splitting and lowercasing
        formattedText = formattedText.lower()
        formattedText = formattedText.split()
        #print('Without uppercases and split:', formattedText)
        
        #maketrans() function is used to construct the translation table
        #specifies the list of characters that need to be replaced or the characters that need to be deleted 
        table = str.maketrans('', '', string.punctuation)
        
        #translate() function uses the translation mapping specified using the maketrans()
        strippedText = [w.translate(table) for w in formattedText]
        #print('Without punctuation:', strippedText)
        
        #remove numbers
        cleanTextNum = [word for word in strippedText if word.isalpha()]
        #print('Without numbers:', cleanTextStop)
        
        #remove stopwords
        cleanTextStop = [word for word in cleanTextNum if not word in all_stopwords]
        #print('Without stopwords:', cleanTextStop)
       
        #join final words back into complete sentences
        finalText = (" ").join(cleanTextStop)
        self.fmtText = finalText
        #assign finalText to self.fmtText variable
        
        #print(finalText)
          
    def freqAll(self):        
        # split text into words
        wordList = self.fmtText.split(' ')
        #print(wordList)
        #print(set(wordList))
        
        # Create dictionary
        freqMap = {}
        
        for word in set(wordList): # use set function to remove duplicates in list
            #print(word, wordList.count(word))
            freqMap[word] = wordList.count(word)
            
        return freqMap
        #print freqMap
    
    def freqOf(self,word):
        # get frequency map
        freqDict = self.freqAll()
        
        if word in freqDict:
            return freqDict[word]
        else:
            #print("Word doesn't exist!")
            return 0
        
    #third function built to return the min and max frequency val and word
    #rarely use this function
    def MinMaxWords(self):
        freqDict2 = self.freqAll()
        max_key = max(freqDict2, key = freqDict2.get)
        min_key = min(freqDict2, key = freqDict2.get)
        
        print('Maximum Value: ',max_key)
        print('Minimum Value: ',min_key)
```

<h2> Calling our Class & Method #1 - Frequency of all Words üó£ </h2>

```python
import sys

#specify where your .txt file is located
filename = '/Users/samdestura/desktop/participant-analysis/participant.txt'
open_file = open(filename, 'r')
file_contents = open_file.readlines()

#pass file_contents throught analysedText class and assign clean text to a variable
output = analysedText(file_contents)

#call freqAll() method using the object of analysedText class aka 'output'
#store freqAll() results in a new variable aka 'result'
#print the results with a helpful text statement
result = output.freqAll()
print('Here is the frequency of each word within your text file:', result)
```
